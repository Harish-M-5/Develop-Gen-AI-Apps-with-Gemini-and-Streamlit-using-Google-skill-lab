```python
import google.generativeai as genai

# Configure API Key
genai.configure(api_key="YOUR_API_KEY")

# Load Gemini model
model = genai.GenerativeModel("gemini-pro")

# Generate response
response = model.generate_content(
    "Explain Generative AI in simple terms"
)

print(response.text)
```
